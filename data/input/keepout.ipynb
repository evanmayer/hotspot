```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ax = plt.axes()
ax.axhline(0, color='k', linewidth=2)
ax.axhline(1, color='k', linewidth=2)
ax.axvline(0, color='k', linewidth=2)
ax.axvline(25, color='k', linewidth=2)

kwargs = {
    'color': 'r',
    'alpha': .2
}
# lower keepouts
y0 = np.arange(0,.11,.01)
ax.fill_betweenx(y0, 1.582, 3.332, **kwargs, label='Keep-out zones')
ax.fill_betweenx(y0, 6.849, 8.599, **kwargs)
ax.fill_betweenx(y0, 16.991, 18.741, **kwargs)
ax.fill_betweenx(y0, 17.923, 19.673, **kwargs)

# upper keepouts
y1 = np.arange(.9,1.01,.01)
ax.fill_betweenx(y1, 8.597, 10.347, **kwargs)
ax.fill_betweenx(y1, 11.920, 13.670, **kwargs)

# plot available holes
spacing = 1.5
[ax.axvline(x, linestyle='--', color='silver', label='Mounting Hole Locations') if i == 0 else ax.axvline(x, linestyle='--', color='silver') for i, x in enumerate(np.arange(1, 25, 1.5))]

ax.grid(True)
ax.legend(loc='best')
ax.set_title('K-Mirror Laser Tracker Nest Keep-Out Zones')
ax.set_ylabel('Mirror Edge y-extent (arb units)')
ax.set_xlabel('Distance from Mirror Reference Edge (in)')
plt.savefig('../../doc/img/keepout.png')
```
